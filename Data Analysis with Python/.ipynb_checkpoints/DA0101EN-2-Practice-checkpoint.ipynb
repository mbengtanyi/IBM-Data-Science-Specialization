```python
#importing modules
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
#loading the data into a dataframe
path = "/Users/tanyi/Documents/GitHub/IBM Data Science Specialization/Data Analysis with Python/imports-85.csv"

#adding the headers since the orignal csv did not have them
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"] 

df = pd.read_csv(path, names=headers)

df.head(20)
```

```python
#Replacing the "?" cells with NaN
df.replace("?", np.nan, inplace=True)
df.head()
```

```python
#Evaluating for missing data using .isnull() and .isnotnull()
missing_data = df.isnull()
missing_data.head()
```

```python
#Counting how many cells have missing values
df.info() #simply!! or...

# for column in missing_data.columns.values.tolist():
#     print(column, missing_data[column].value_counts())
#     #print(missing_data[column].value_counts())
#     print("")
```

Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely.
We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

<b>Replace by mean:</b>

<ul>
    <li>"normalized-losses": 41 missing data, replace them with mean</li>
    <li>"stroke": 4 missing data, replace them with mean</li>
    <li>"bore": 4 missing data, replace them with mean</li>
    <li>"horsepower": 2 missing data, replace them with mean</li>
    <li>"peak-rpm": 2 missing data, replace them with mean</li>
</ul>

<b>Replace by frequency:</b>

<ul>
    <li>"num-of-doors": 2 missing data, replace them with "four". 
        <ul>
            <li>Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur</li>
        </ul>
    </li>
</ul>

<b>Drop the whole row:</b>

<ul>
    <li>"price": 4 missing data, simply delete the whole row
        <ul>
            <li>Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us</li>
        </ul>
    </li>
</ul>

```python
#Calculating the mean for normalized losses
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Normalized losses mean:", avg_norm_loss)

#replacing empty values with the mean
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
df.head()
```

```python
#Replace NaN with the mean value in the "bore" column
avg_bore = df["bore"].astype("float").mean(axis=0)
print("Average bore:", avg_bore)
df["bore"].replace(np.nan, avg_bore, inplace=True)
df["bore"].tail(10)
```

```python
#Replacing NaN with the mean value in "horsepower" column
avg_hp = df["horsepower"].astype("float").mean(axis=0)
print("Average horsepower:", avg_hp)

df["horsepower"].replace(np.nan, avg_hp, inplace=True)
df["horsepower"]
```

```python
#Replacing NaN values for peakrpm with the mean
avg_prpm = df["peak-rpm"].astype("float").mean(axis=0)
print("Average peak rpm:", avg_prpm)

df["peak-rpm"].replace(np.nan, avg_prpm, inplace=True)
df["peak-rpm"]
```

```python
df['num-of-doors'].value_counts()
```

```python
#most common value in column 'num-of-doors'
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
#df['num-of-doors']
```

```python
#Dropping rows without data on price
df.dropna(subset=["price"], axis=0, inplace=True)

#resetting index since rows were dropped
df.reset_index(drop=True, inplace=True)
df.info()
```

```python
#Converting columns to more proper data types
df[['bore', 'stroke', 'price', 'peak-rpm']] = df[['bore', 'stroke', 'price', 'peak-rpm']].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
df.dtypes
```
